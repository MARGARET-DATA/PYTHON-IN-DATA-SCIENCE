```python
all_num = range(1,21)
even_list = []
odd_list = []
for i in all_num:
    if i%2==0:
        even_list.append(i)
    else:
            odd_list.append(i)
print(even_list)   
print(odd_list)
```

```python
all_num = range(1,21)
even_list = []
odd_list = []
for i in all_num:
    if i==14:
        break
    if i%2==0:
        even_list.append(i)
    else:
            odd_list.append(i)
print(even_list)   
print(odd_list)
```

```python
all_num = range(1,21)
even_list = []
odd_list = []
for i in all_num:
    if i==14:
         continue
    if i%2==0:
        even_list.append(i)
    else:
            odd_list.append(i)
print(even_list)   
print(odd_list)
```
